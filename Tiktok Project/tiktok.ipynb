```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler , RobustScaler , MinMaxScaler , PolynomialFeatures , LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB , MultinomialNB , BernoulliNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier , plot_tree
from sklearn.metrics import accuracy_score , recall_score , precision_score , f1_score , confusion_matrix , ConfusionMatrixDisplay
```

```python
ohe = OneHotEncoder(sparse_output=False , drop = 'first')

col = ['verified_status', 'author_ban_status']


df = pd.concat([df , pd.DataFrame(ohe.fit_transform(df[col]) , columns=ohe.get_feature_names_out())] , axis = 1)

df.drop(col , axis = 1 , inplace = True )
```

```python
le = LabelEncoder()

df['claim_status'] = le.fit_transform(df['claim_status'])
```

```python
x , y = df.drop(['claim_status'] , axis = 1 ) , df['claim_status']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=7 , stratify=y)
```

```python
sclr = StandardScaler()

x_train = sclr.fit_transform(x_train)
x_test = sclr.transform(x_test)
```

```python
poly = PolynomialFeatures()
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
def Modeling (model , x_train , x_test , y_train , y_test):
    
    model.fit(x_train , y_train)
    
    y_train_prd = model.predict(x_train)
    y_test_prd = model.predict(x_test)
    
    print(f'Train acc : {accuracy_score(y_train , y_train_prd)}')
    print(f'Test acc : {accuracy_score(y_test , y_test_prd)}')
    print('*'*50)
    print(f'Train recall_score : {recall_score(y_train , y_train_prd)}')
    print(f'Test recall_score : {recall_score(y_test , y_test_prd)}')
    print('*'*50)
    print(f'Train precision_score : {precision_score(y_train , y_train_prd)}')
    print(f'Test precision_score : {precision_score(y_test , y_test_prd)}')
    print('*'*50)
    print(f'Train f1_score : {f1_score(y_train , y_train_prd)}')
    print(f'Test f1_score : {f1_score(y_test , y_test_prd)}')
    print('*'*50)
    
    ConfusionMatrixDisplay(confusion_matrix(y_train , y_train_prd)).plot()
    ConfusionMatrixDisplay(confusion_matrix(y_test , y_test_prd)).plot()
```

```python
Modeling(LogisticRegression() , x_train , x_test , y_train , y_test)
```

```python
Modeling(LogisticRegression(max_iter=5000) , x_train_poly , x_test_poly , y_train , y_test)
```

```python
Modeling(KNeighborsClassifier(n_neighbors=7) , x_train , x_test , y_train , y_test)
```

```python
Modeling(GaussianNB() , x_train , x_test , y_train , y_test)
```

```python
Modeling(SVC(kernel='linear' , C = 100) , x_train , x_test , y_train , y_test)
```

```python
Modeling(DecisionTreeClassifier(max_depth=20 , min_samples_split=20) , x_train , x_test , y_train , y_test)
```
