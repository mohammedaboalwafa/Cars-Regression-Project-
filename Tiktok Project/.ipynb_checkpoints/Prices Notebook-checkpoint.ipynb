```python
# Drop Unneccessary columns before preprocessing

df.drop(['date'], axis= 1, inplace= True)

df.duplicated().sum()
```

```python
df
```

```python
# Split data into inpute features and target variable

x = df.drop('price', axis= 1)
y = df['price']
```

```python
# split into train and test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 42)
x_train
```

### Feature scaling

```python
scaled_cols = ['bedrooms', 'sqft_lot', 'yr_built', 'yr_renovated', 'day','sqft_living', 'sqft_above', 'sqft_basement']

from sklearn.preprocessing import RobustScaler

rc = RobustScaler()
x_train[scaled_cols] = rc.fit_transform(x_train[scaled_cols])
x_test[scaled_cols] = rc.transform(x_test[scaled_cols])
```

```python
x_train
```

### Handling Categorical Features (Nominal)

```python
x_train
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop= 'first', sparse_output= False)

x_train['month'] = ohe.fit_transform(x_train[['month']])
x_test['month'] = ohe.transform(x_test[['month']])
```

```python
x_train.month
```

```python
# Features with high categories, handle by Binary encoder
from category_encoders import BinaryEncoder
BE = BinaryEncoder()

x_train_BE = BE.fit_transform(x_train[['city']])
x_test_BE = BE.transform(x_test[['city']])
```

```python
x_train_BE
```

```python
x_train.drop('city', axis= 1, inplace= True)
x_test.drop('city', axis= 1, inplace= True)
```

```python
x_train = pd.concat([x_train, x_train_BE], axis= 1)
x_test = pd.concat([x_test, x_test_BE], axis= 1)
```

```python
x_train
```

### Handling Categorical Features (ordinal)

```python
x_train.bathrooms.value_counts()
```

```python
# Arrange each category as per importance
bathrooms_map = {'0.75 - 2' : 1,
                 '2.25 - 3' : 2,
                 '3.25 - 4' : 3,
                 '4.25 - 8' : 4}

x_train.bathrooms = x_train.bathrooms.map(bathrooms_map)
x_test.bathrooms = x_test.bathrooms.map(bathrooms_map)
```

```python
x_train.bathrooms
```

```python
x_train.floors.value_counts()
```

```python
# Arrange each category as per importance
floors_map = {'1 - 1.5' : 1,
                 '2 - 2.5' : 2,
                 '3 - 3.5' : 3
                 }

x_train.floors = x_train.floors.map(floors_map)
x_test.floors = x_test.floors.map(floors_map)
```

```python
x_train.floors
```

```python
x_train
```

### Machine Learning

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge

le = LinearRegression()

le.fit(x_train, np.log(y_train))

print('Training score : ', round(le.score(x_train, np.log(y_train)) * 100, 2), '%')
print('Test score : ', round(le.score(x_test, np.log(y_test)) * 100, 2), '%')
```

```python
la = Lasso(0.0001)

la.fit(x_train, np.log(y_train))
print('Training score : ', round(la.score(x_train, np.log(y_train)) * 100, 2), '%')
print('Test score : ', round(la.score(x_test, np.log(y_test)) * 100, 2), '%')
```

```python
rd = Ridge()

rd.fit(x_train, np.log(y_train))
print('Training score : ', round(rd.score(x_train, np.log(y_train)) * 100, 2), '%')
print('Test score : ', round(rd.score(x_test, np.log(y_test)) * 100, 2), '%')
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree= 2)

x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
le.fit(x_train_poly, np.log(y_train))

print('Training score : ', round(le.score(x_train_poly, np.log(y_train)) * 100, 2), '%')
print('Test score : ', round(le.score(x_test_poly, np.log(y_test)) * 100, 2), '%')
```
