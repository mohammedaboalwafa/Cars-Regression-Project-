```python
be_col = ['location', 'brand']
# be_col = ['location']
ohe_col = ['fuel_type' , 'transmission']
```

#### Data Preparation 

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler , RobustScaler , MinMaxScaler , PolynomialFeatures
from category_encoders import BinaryEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression , Ridge , Lasso
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
```

```python
ohe = OneHotEncoder(sparse_output=False , drop = 'first')

df_ohe = pd.DataFrame(data = ohe.fit_transform(df[ohe_col]) , columns=ohe.get_feature_names_out())

df = pd.concat([df , df_ohe] , axis = 1 )
```

```python
be = BinaryEncoder()

df_be = be.fit_transform(df[be_col])

df = pd.concat([df , df_be] , axis = 1 )
```

```python
be_col + ohe_col
```

```python
df.drop(be_col + ohe_col , axis = 1 , inplace = True )
```

```python
x , y = df.drop(['price'] , axis = 1 ) , np.log(df['price'])
# x , y = df.drop(['price'] , axis = 1 ) , df['price']
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.20 , random_state=7)
```

```python
sclr = RobustScaler()

x_train[x_train.columns] = sclr.fit_transform(x_train)

x_test[x_test.columns] = sclr.transform(x_test)
```

```python
poly = PolynomialFeatures(degree=2)

x_train_poly = poly.fit_transform(x_train)

x_test_poly = poly.transform(x_test)
```

```python
x_train.head()
```

#### lr without poly

```python
lr = LinearRegression()

lr.fit(x_train , y_train)

y_train_prd = lr.predict(x_train)
y_test_prd = lr.predict(x_test)

print(f'Train score : {r2_score(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test score : {r2_score(y_true= y_test , y_pred=y_test_prd)}')

print('*'*30)

print(f'Train MSE : {mean_squared_error(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test MSE : {mean_squared_error(y_true= y_test , y_pred=y_test_prd)}')
```

#### lr with poly

```python
lr = LinearRegression()

lr.fit(x_train_poly , y_train)

y_train_prd = lr.predict(x_train_poly)
y_test_prd = lr.predict(x_test_poly)

print(f'Train score : {r2_score(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test score : {r2_score(y_true= y_test , y_pred=y_test_prd)}')

print('*'*30)

print(f'Train MSE : {mean_squared_error(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test MSE : {mean_squared_error(y_true= y_test , y_pred=y_test_prd)}')
```

#### lasso without poly 

```python
lasso = Lasso(alpha=1)

lasso.fit(x_train , y_train)

y_train_prd = lasso.predict(x_train)
y_test_prd = lasso.predict(x_test)

print(f'Train score : {r2_score(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test score : {r2_score(y_true= y_test , y_pred=y_test_prd)}')

print('*'*30)

print(f'Train MSE : {mean_squared_error(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test MSE : {mean_squared_error(y_true= y_test , y_pred=y_test_prd)}')
```

#### lasso with poly

```python
lasso = Lasso(alpha=1)

lasso.fit(x_train_poly , y_train)

y_train_prd = lasso.predict(x_train_poly)
y_test_prd = lasso.predict(x_test_poly)

print(f'Train score : {r2_score(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test score : {r2_score(y_true= y_test , y_pred=y_test_prd)}')

print('*'*30)

print(f'Train MSE : {mean_squared_error(y_true= y_train , y_pred=y_train_prd)}')
print(f'Test MSE : {mean_squared_error(y_true= y_test , y_pred=y_test_prd)}')
```
